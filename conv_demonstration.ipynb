## Conv1D understanding

```python
conv_1d = nn.Conv1d(in_channels=3, out_channels=1, kernel_size=(1,1))
for x in conv_1d.parameters():
    print(x.shape, x.numel())
```

## counting the number of parameters in Conv2D layer

```python
l = nn.Conv2d(in_channels=3, out_channels=64, kernel_size=7, stride=2, padding=3)
total_params = sum(p.numel() for p in l.parameters())
```

```python
for x in l.parameters():
    print(x.shape)
```
